```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

#data = pd.read_csv("cancer_outlier.csv") #Loading Breast Cancer data
data = pd.read_csv("cancer.csv") #Loading Breast Cancer data

X = data.iloc[:,0] # First column of data
Y = data.iloc[:,1] # Target variable i.e. if tumor is benign or malignant

y = np.array(Y) #Converting pandas object Y to a numpy array 
X = np.array(X) #Converting pandas object X to a numpy array 
print(X.shape)

# Encoding B as 0, and M as 1
for i in range(len(y)):
    if(y[i]=='M'):
        y[i]=1
    else:
        y[i]=0      
   
data['Type'] = y

sns.lmplot(x="Size", y="Type", data=data, hue="Tumor", fit_reg=False, legend=False)

plt.title('Cancer Dataset Distribution')
plt.legend()
plt.show()
        
# Splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=13) 

# Converting data from 1D to 2D
X_train = X_train.reshape((-1,1))
X_test = X_test.reshape((-1,1))

lr = linear_model.LinearRegression()
lr.fit(X_train, y_train) # Fitting data on linear regression

y_predict = lr.predict(X_test) # Predicting results on test data

X = X.reshape((-1,1))

plt.scatter(X, y)
plt.plot(X, lr.predict(X))
plt.title('Regression Line on Cancer Data')
plt.xlabel('Size')
plt.ylabel('Type')
plt.show()


y_predict_full = lr.predict(X)
y_predict_str = [None]*len(y_predict_full)

print("coefficient:", lr.coef_) # Coefficients
print("Intercept:",lr.intercept_) # Intercept/Bias
print("MSE: ",mean_squared_error(y_test, y_predict)) # Mean Squared Error
print("R2_score:", r2_score(y_test, y_predict)) # R2 score

print("Min value of y predicted:",min(y_predict_full))
print("Min value of y predicted:",max(y_predict_full))
      
threshold = 0.5 # Threshold value for prediction
for i in range(len(y_predict_full)):
    if(y_predict_full[i]<=threshold):
        y_predict_str[i]='B'
    else:
        y_predict_str[i]='M'


data['predicted'] =  y_predict_str
#data['actual'] = y

sns.lmplot(x="Size", y="Type", data=data, hue="predicted", fit_reg=False, legend=False)

plt.title('Linear Regression on Cancer Dataset')
plt.legend()
plt.show()
```
